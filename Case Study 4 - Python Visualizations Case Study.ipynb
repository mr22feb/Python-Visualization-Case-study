```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sales = pd.read_csv('Downloads/Case Study 4 - Python Visualizations Case Study/SalesData.csv')
```

```python
sales.head(20)
```

#### 1. Compare Sales by region for 2016 with 2015 using bar chart


```python
sale = sales.groupby(by=('Region')).agg({'Sales2015':'sum','Sales2016':'sum'}).round(2)

sale.plot(kind='bar')
plt.title('Sales by region 2015 & 2016')
plt.ylabel('Sales')
plt.show()
```

#### 2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.

```python
sales_pie = sales.groupby(by='Region')['Sales2016'].sum()

sales_pie
```

```python
plt.pie(sales_pie,autopct="%1.0f%%",labels=["Central","East","West"],shadow=True,explode=[0.0,0.1,0.0]
                                                                                ,colors=['r', 'g', 'b'])
plt.title("Contributing factors to the sales for each region in 2016")
plt.show()
```

#### 3. Compare the total sales of 2015 and 2016 with respect to Region and Tiers


```python
region_wise = sales.groupby(by=['Region','Tier'])['Sales2015','Sales2016'].sum()


region_wise.plot(kind='barh',stacked=True,figsize=(8,8))
plt.ylabel("Sales")
plt.title("Total sales of 2015 and 2016 with respect to Region and Tiers")
plt.show()
```

```python
print("""We can conclude that East region and High tier in 2016 have contributed the maximum. 
Also sales in 2015 and sales in 2016 both is maximum in East region and High tier""")
```

#### 4. In East region, which state registered a decline in 2016 as compared to 2015?


```python
sales_state = sales.loc[sales['Region']=='East',:].groupby(['Region',"State"])['Sales2015',
                                                          'Sales2016'].sum()

sales_state.plot(kind='bar',figsize=(10,8))
plt.ylabel("Sales")
plt.title("Sales comparison between 2015 and 2016 for East Region")
plt.show()
print("NY state registered a decline in sales in 2016 as compared to 2015")
```

#### 5. In all the High tier, which Division saw a decline in number of units sold in 2016 compared to 2015?


```python
high_tier = sales.loc[sales['Tier']=='High',:].groupby(by='Division')['Units2015','Units2016'].sum()

high_tier.plot(kind='barh',figsize=(8,10))
plt.show()
print("No division show decline in number of units sold in 2016 compared to 2015")
```

#### 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported 
#### dataset. The Quarters are based on months and defined as -
####  Jan - Mar : Q1
####  Apr - Jun : Q2
####  Jul - Sep : Q3
####  Oct - Dec : Q4 

```python
sales['Quarter'] = np.where(sales['Month'].isin(['Jan', 'Feb', 'Mar']), 'Q1',
                    np.where(sales['Month'].isin(['Apr', 'May', 'Jun']), 'Q2',
                    np.where(sales['Month'].isin(['Jul', 'Aug', 'Sep']), 'Q3',
                    np.where(sales['Month'].isin(['Oct', 'Nov', 'Dec']), 'Q4', 'Unknown'))))

sales.head()
```

#### 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot
 

```python
qty_sales = sales.groupby(by=('Quarter'))['Sales2015','Sales2016'].sum()

qty_sales.plot(kind='bar',figsize=(10,5))
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```

#### 8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart.
#### (Draw 4 pie charts representing a Quarter for each Tier)


```python
qtr_pivot = sales.pivot_table(index='Quarter',columns='Tier',values='Sales2016')
qtr_pivot
```

```python
#Qtr Q1
plt.pie(x=qtr_pivot.loc["Q1",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],
                                colors=['deepskyblue', 'pink', 'orange'])
plt.show()
```

```python
#Qtr Q2
plt.pie(x=qtr_pivot.loc["Q2",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],
                                  colors=['deepskyblue', 'pink', 'orange'])
plt.show()
```

```python
#Qtr Q3

q3 = qtr_pivot.loc["Q3", :]
q3_abs = q3.abs()
plt.pie(x=q3_abs, autopct="%1.0f%%", labels=["High", "Low", "Med", "Out"],
        colors=['deepskyblue', 'pink', 'orange'])
plt.show()
```

```python
#Qtr Q4

plt.pie(x=qtr_pivot.loc["Q4",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"]
        ,colors=['deepskyblue', 'pink', 'orange'])
plt.show()
```
